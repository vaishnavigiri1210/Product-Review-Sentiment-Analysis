```python
import pandas as pd     # data handling
```

```python
# data loading

df_1 = pd.read_csv('../data/flipkart_product.csv', encoding='latin-1')
df_2 = pd.read_csv('../data/Reviews.csv', encoding='utf-8-sig')
```

```python
print(df_1.shape)       # (number of rows, number of columns)
print(df_2.shape)
```

```python
# cleaning column names by removing unwanted characters and stripping whitespace
# str.replace => replaces specified characters with an empty string
# str.strip => removes leading and trailing whitespace from the column names

df_1.columns = df_1.columns.str.replace('ï»¿', '').str.replace('Ã¯Â»Â¿', '').str.strip()
df_2.columns = df_2.columns.str.replace('ï»¿', '').str.strip()
```

```python
print("Cleaned Columns df1:", df_1.columns.tolist())
print("Cleaned Columns df2:", df_2.columns.tolist())
```

```python
print(df_1.head())      # first 5 rows of the dataframe
print(df_2.head())
```

```python
# selecting only the 'Review' and 'rating' columns and renaming them to 'review_text' and 'rating' respectively for df_1

df_1_clean = df_1[['Review', 'rating']].rename(
    columns={'Review': 'review_text', 'rating': 'rating'}
)
```

```python
# selecting only the 'Review' and 'rating' columns and renaming them to 'review_text' and 'rating' respectively for df_1

df_2_clean = df_2[['Review', 'rating']].rename(
    columns={'Review': 'review_text', 'rating': 'rating'}
)
```

```python
# combining the cleaned dataframes df_1_clean and df_2_clean into a single dataframe df_combined using pd.concat

df_combined = pd.concat([df_1_clean, df_2_clean], ignore_index=True)
```

```python
# dropping rows with missing values in 'review_text' and 'rating' columns
# converting 'rating' to numeric and dropping rows where conversion fails
# dropna => removes rows with missing values in specified columns
# to_numeric => converts values to numeric, with errors='coerce' converting non-convertible values to NaN
# inplace=True => modifies the dataframe in place without creating a new one

df_combined.dropna(subset=['review_text', 'rating'], inplace=True)
df_combined['rating'] = pd.to_numeric(df_combined['rating'], errors='coerce')
df_combined.dropna(subset=['rating'], inplace=True)
```

```python
# shuffling the combined dataframe df_combined to ensure that the reviews from both datasets are mixed well, and resetting the index after shuffling

df_combined = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# stripping leading and trailing whitespace from the 'review_text' column to ensure that the review texts are clean and do not have unnecessary spaces
df_combined['review_text'] = df_combined['review_text'].str.strip()
```

```python
# filtering out reviews that are too short (length <= 1) as they may not provide meaningful information for sentiment analysis
df_combined = df_combined[df_combined['review_text'].str.len() > 1]
```

```python
# checking for duplicates in the combined dataframe df_combined and printing the total number of duplicates found
duplicates = df_combined.duplicated().sum()
print(f"Total duplicates found: {duplicates}")
```

```python
# saving the cleaned and combined dataframe df_combined to a new CSV file named 'cleaned_data.csv' in the '../data/' directory without the index and using 'utf-8-sig' encoding to ensure proper handling of special characters
df_combined.to_csv('../data/cleaned_data.csv', index=False, encoding='utf-8-sig')
```
